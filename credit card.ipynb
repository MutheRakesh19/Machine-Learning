```python
# Step 1: Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Step 2: Generate or load data
# For demonstration, let's create some synthetic data
X, _ = make_blobs(n_samples=300, centers=2, cluster_std=0.60, random_state=0)

# Step 3: Calculate WCSS for different numbers of clusters
wcss = []
for i in range(1, 11):  # Number of clusters from 1 to 10
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # Inertia is another term for WCSS

# Step 4: Plot the elbow graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2,random_state=42)
clusters = kmeans.fit_predict(df)
df['cluster'] = clusters
```

```python
pca = PCA (n_components = 2)
reduced_df = pca.fit_transform(df)
```

```python
plot_df = pd.DataFrame(reduced_df,columns =['pc1','pc2'])
plot_df['cluster'] = clusters
```

```python
plt.figure(figsize=(10, 8))
scatter = plt.scatter(plot_df['pc1'], plot_df['pc2'], c=plot_df['cluster'], cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering Visualization')
plt.show()
```

```python
cluster_counts = df['cluster'].value_counts()
print('cluster_counts: ')
print(cluster_counts)
```

```python
threshold = cluster_counts.median()

high_clusters = cluster_counts[cluster_counts > threshold].index
low_clusters = cluster_counts[cluster_counts <= threshold].index

            
```

```python
high_data =df[df['cluster'].isin(high_clusters)]
low_data = df[df['cluster'].isin(low_clusters)]
```
