```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q1-q3
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
    
```

```python
wisker(df[col])
```

```python
for i in ['Age', 'Height', 'Weight', 'NCP']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
    
```
